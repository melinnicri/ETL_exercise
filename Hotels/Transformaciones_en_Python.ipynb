```python
# Replace nulls with zeros
dfHotelBooking.fillna(0, inplace=True)
```

```python
dfHotelBooking.isnull().sum()
```

```python
# The data type is changed
dfHotelBooking['children'] = dfHotelBooking['children'].values.astype(int)
dfHotelBooking['adults'] = dfHotelBooking['adults'].values.astype(int)
dfHotelBooking['babies'] = dfHotelBooking['babies'].values.astype(int)
```

```python
# A new column is created
# the total of people column is created, which will be the sum of the amount of each group of people
dfHotelBooking['Amount of people'] = dfHotelBooking['babies'] + dfHotelBooking['children'] + dfHotelBooking['adults']
```

```python
dfHotelBooking["market_segment"].unique()
```

```python
# Replace one value with another
# The term “TA” is “Travel Agents” and “TO” is “Tour Operators”
dfHotelBooking["market_segment"].replace({"Online TA": "Online Travel Agents", "Offline TA/TO": "Off Travel Agents/Tour Operators "}, inplace=True)
```

```python
dfHotelBooking
```

```python
# The names of the columns (headers) are taken to pass them through the melt
data_top = list(dfHotelBooking.columns.values)

data_top
```

```python
# The adults, children and babies column is eliminated since these are the ones that will be melted
data_top.remove('adults')
data_top.remove('children')
data_top.remove('babies')
# pandas.melt is used to reorganize the dataframe structure
dfHotelBooking=dfHotelBooking.melt(data_top, var_name='A group of people')
```

```python
dfHotelBooking
```

```python
# The column name is changed
dfHotelBooking= dfHotelBooking.rename(columns = {'value':'Amount'}, inplace = False)
```

```python
# The drop_duplicates() function returns a dataframe with duplicate rows removed
dfHotelBooking.drop_duplicates(inplace=True)
```

```python
# Filter columns, leaving only the values that have 2015 in the arrival year column
dfHotelBooking = dfHotelBooking.loc[dfHotelBooking['arrival_date_year'] == 2015]
```

```python
# It is verified that the filter was performed, now there are fewer columns
dfHotelBooking.info()
```

```python
# Create a column with random numbers or values, which are delivered in an array
dfHotelBooking['randNumCol'] = np.random.choice(['cool', 9, 20], dfHotelBooking.shape[0])
```

```python
# Applying multiple aggregation functions to a grouping of data made with group by
# the numerical value that the aggregation functions will use must be added to the list, in this case it is quantity
dfHotelBooking = dfHotelBooking.query('country != 0')
dfHotelBooking = dfHotelBooking[['country','A group of people','Amount']].groupby(['country','A group of people']).agg(['mean','count'])
dfHotelBooking.head()
```

```python
# The columns are organized again, eliminating an index level and resetting it
# a level is removed
dfHotelBooking.columns = dfHotelBooking.columns.droplevel()
```

```python
# The index is reset
dfHotelBooking = dfHotelBooking.reset_index()
```

```python
# You can generate a column with a ranking, the method is min so the smallest value will have the first place
# is applied by the mean column
dfHotelBooking['rank'] = dfHotelBooking['mean'].rank(method = 'min')
```

```python
# You can sort the values with the sort_values function
dfHotelBooking.sort_values('rank',inplace = True)
```

```python
# With the sample function you can take a sample of the dataset, with a specific number of rows
dfHotelBooking = dfHotelBooking.sample(n=100)
```
